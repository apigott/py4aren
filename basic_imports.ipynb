What can we do with numpy?

```python
x = np.arange(0,10,0.5) # creates a numpy array from 0-10 at intervals of 0.5
y1 = np.sin(x) # does the element-wise function sin() on all values of x
y2 = np.cos(x)
print(y1)
```

If we just type the name of a variable, Jupyter will print out the variable for us. This is a convinient alternative to print().

Often we want to plot stuff. Matplotlib is a common package for plotting:

```python
# pip install matplotlib # uncomment this if you haven't installed it yet!
```

```python
import matplotlib.pyplot as plt # i only want the "pyplot" module from matplotlib
plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos")
plt.legend()
```

```python
Here they are as subplots
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(y1) #  what's the difference in the x axis here?
ax[1].plot(y2, color='red') # if you take away the color arg, they will both be blue
```
